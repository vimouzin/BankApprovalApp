```python
import yaml
import io
from tamr_unify_client import Client
from tamr_unify_client.auth import UsernamePasswordAuth
import json
from kafka import KafkaConsumer
import pandas as pd
import requests
import requests, json



def readConfig(): # Read YAML file
    with open("data.yaml", 'r') as stream:
     data_loaded = yaml.safe_load(stream)
     return data_loaded

def old_llm(query_data, config_data):
   df_clusterlist = get_cluster_list(query_data,config_data)
   response = golden_record_query(df_clusterlist,config_data)
   return response



def get_cluster_list(record,config_data):
       
       request =[{'recordId': 0, 'record':record}]  
       auth = UsernamePasswordAuth(config_data["USERNAME"], config_data["PASSWORD"])
       tamr = Client(auth, host=config_data["IP_ADDRESS"],port=9170,base_path="api/v1")
       project_name = config_data["PROJECT_NAME"]
       response = tamr.post(f"projects/{project_name}:match?",
                params={"type": "clusters"},
                verify=False,
                data="\n".join(json.dumps(i) for i in request).replace("NaN", "null")
                ).successful()

       df_clusterlist = pd.read_json(
            response.text,
            lines=True,
            orient="records",
            typ="frame",
            encoding="utf-8",
            dtype=object,
        )
       df_clusterlist = df_clusterlist.reset_index()
       print(df_clusterlist)
    # make sure indexes pair with number of rows
       return df_clusterlist

def golden_record_query(full_clusterlist,config_data):
     auth = UsernamePasswordAuth(config_data["USERNAME"], config_data["PASSWORD"],)
     tamr = Client(auth, host=config_data["IP_ADDRESS"],port=9100)
     clusters = []
     clusters_scores={}
     for index, row in full_clusterlist.iterrows():
         request ={"persistentId":row['clusterId']}
         clusters_scores[row['clusterId']] = '{0:.2f}'.format(row["avgMatchProb"],2)
         clusters.append(request)
    
     
     response = tamr.post(f'/api/dataset/datasets/{config_data["GR_DT_DATASET_NAME"]}/records',
                            data ="\n".join([json.dumps(x) for x in clusters])) 
     if not response.text:
         return []
     result = [json.loads(i) for i in response.text.split('\n')]
     finalresults =[]
     print(response.text)

     for record in result:
         print(record)
         formated_results={}
         if "persistentId" in record:
            formated_results["persistentId"]= record["persistentId"]
         for cols in config_data["INPUT_ATTRS"]:

             view = cols["ATTR_NAME"]
             value = cols["ATTR_NAME"]
             if value in record:
                formated_results[view] = record[value]
         if "persistentId" in record:
           
            formated_results["Score"] = clusters_scores[record["persistentId"]]
         findConfidence(formated_results)
         finalresults.append(formated_results)
       #  print(formated_results)
         
     return finalresults


def findConfidence(record={}):
    if "Score" in record:
    
        if float(record["Score"]) < 0.4:
            record['Confidence'] = "low"
            record['fillColor'] ="red"

        elif float(record["Score"]) < 0.7:
            record['Confidence'] = "Medium"
            record['fillColor'] ="yellow"

        else:
            record['Confidence'] = "High"
            record['fillColor'] ="green"

def get_source_records(clusterlist,configfile):
     auth = UsernamePasswordAuth(configfile["USERNAME"], configfile["PASSWORD"])
     tamr_client = Client(auth, host=configfile["IP_ADDRESS"])
     tamr_project = tamr_client.projects.by_name(configfile["PROJECT_NAME"])
     response = tamr_client.get(
                        "/api/dedup/records/{0}?clusterIds={1}".format(
                            tamr_project.unified_dataset().name, "&clusterIds="+clusterlist
                        ),
                        verify=False,
                    ).successful().json()
              
     source_records =[]
     print(configfile)
     for record in response["items"]:
        record_to_add ={}
        print(record)
        for attr in configfile["INPUT_ATTRS"]:
            if "tamr__"+attr["ATTR_NAME"] in record["source"]:
           

             if hasattr(record["source"]["tamr__"+attr["ATTR_NAME"]], '__len__') and (not isinstance(record["source"]["tamr__"+attr["ATTR_NAME"]], str)):
            
               if len(record["source"]["tamr__"+attr["ATTR_NAME"]]) >0:

                   
               
                    record_to_add[attr["ATTR_VIEW"]] = record["source"]["tamr__"+attr["ATTR_NAME"]][0]
        record_to_add["Source Dataset"] = record["source"]["origin_source_name"]
        source_records.append(record_to_add)
     return source_records
    
headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}


m = {'Customer': 'lexi Stevenson', 'lastName': 'Stevenson', 'ZIP': '', 'ADDRESS': '9 South 1st Lane', 'STATE': '', 'CITY': '', 'street': '', 'income': '', 'amount': ''}

response = requests.post("http://10.20.1.201:3000/test", data=json.dumps(old_llm(json.dumps(m),readConfig())),headers=headers)
print(response.json())
```
